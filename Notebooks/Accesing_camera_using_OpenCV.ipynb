```python
import cv2
import sys
import os

s = 0

# If running inside Jupyter Notebook, ignore sys.argv completely
if 'ipykernel' in sys.modules:
    s = 0
else:
    if len(sys.argv) > 1:
        arg = sys.argv[1]
        # Only accept numeric values as camera index
        if arg.lstrip("-").isdigit():
            s = int(arg)
        else:
            s = 0

source = cv2.VideoCapture(s)
win_name = "Camera Preview"
cv2.namedWindow(win_name, cv2.WINDOW_NORMAL)

print("Opening source:", s)

if not source.isOpened():
    print("ERROR: Cannot open video source:", s)
    raise SystemExit

while True:
    has_frame, frame = source.read()
    if not has_frame:
        print("No frame received.")
        break

    cv2.imshow(win_name, frame)

    key = cv2.waitKey(1) & 0xFF
    if key == 27 or key == ord('q'):   # ESC or q
        break

source.release()
cv2.destroyAllWindows()
```
